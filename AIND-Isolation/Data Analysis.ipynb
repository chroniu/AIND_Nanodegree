```python
import re
import numpy as np
import pandas as pd

print("Reading file")
data = ""
with open('experiments.txt', 'r') as myfile:
    data = myfile.read()
print("finish reading file")


def get_all(data, pattern):
    result = []
    for line in data.split("\n"):
        if (" "+pattern+" ") in line:
            experiment = []
            numbers = re.split(' |, ', line)
            numbers = list(filter(None, numbers)) 
            for i in range(2, len(numbers)):                
                if(numbers[i].isnumeric()):
                    experiment.append(numbers[i])
            if(experiment != []):
                result.append(experiment)
    return result

        
```

```python

patterns = ["Random", "MM_Open", "MM_Center", "MM_Custom", "MM_Custom_2",
            "MM_Improved", "AB_Open", "AB_Center", "AB_Custom", "AB_Custom_2",
            "AB_Improved"]

columns = np.array(["AB_Improved_W", "AB_Improved_L",
           "AB_Open_W",     "AB_Open_L",
           "AB_Custom_W",   "AB_Custom_L",
           "AB_Custom_2_W", "AB_Custom_2_L",
            "AB_Custom_3_W","AB_Custom_3_L",
           "AB_Custom_4_W", "AB_Custom_4_L" ])

#dfs = pd.DataFrame(index=patterns, columns=columns)
#print(dfs)
#index = 0
dfs = []
for pattern in patterns:
    exp = np.array(get_all(data, pattern)).astype(int)
    #index += 1
    #df = pd.DataFrame(exp)
    #df.columns = columns
    #dfs.append(df.sum().transpose())
    dfs.append(exp.std(axis=0))

dfs = pd.DataFrame(np.array(dfs), columns = columns, index=patterns)

#dfs = pd.concat(dfs)
```
